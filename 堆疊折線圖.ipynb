```python
import matplotlib.pyplot as plt
x = range(0,10)
y1 = [1,5,10,6,9,20,18,19,22,12]
y2 = [12,5,8,9,18,12,20,25,16,7]
y3 = [5,9,14,18,21,4,7,19,2,5]
plt.stackplot(x,y1,y2,y3)
plt.show()
```

```python
x = range(0,10)
y1 = [1,5,10,6,9,20,18,19,22,12]
y2 = [12,5,8,9,18,12,20,25,16,7]
y3 = [5,9,14,18,21,4,7,19,2,5]
fig, ax = plt.subplots(2,2)
fig.set_size_inches(12,8)
ax[0,0].stackplot(x,y1,y2,y3,baseline='zero')
ax[0,0].set_xlabel("zero")
ax[0,1].stackplot(x,y1,y2,y3,baseline='sym')
ax[0,1].set_xlabel("sym")
ax[1,0].stackplot(x,y1,y2,y3,baseline='wiggle')
ax[1,0].set_xlabel("wiggle")
ax[1,1].stackplot(x,y1,y2,y3,labels=['y1','y2','y3'],baseline='weighted_wiggle',colors=['#f66','#0a4','#fa0'])
ax[1,1].set_xlabel("wiggle")
ax[1,1].legend(loc='lower left')
plt.show()
```
